# TOPIC (01): CREATING A DATA FRAME(2-D ARRAY)

```python
import pandas as pd
import numpy as np
```

## 1.1:- USING ARRAYS

```python
#creating an array and storing in DATA variable
data = np.array([[1,2],[3,4],[5,6]])
```

```python
# creating a dataframe using the array created
DF = pd.DataFrame(data, index=['row1','row2','row3'],columns=['col1','col2'])
# index is defining the name of the rows and names have to be stored  inside the list and same goes for columns too.
# this DataFrame is being stored in the variable DF and hence when printing then it will show the DataFrame.
```

```python
# showing the DataFrame and don't PRINT it
DF
```

## 1.2:- USING LISTS

```python
#creating an array and storing in DATA variable
data = ([[1,2],[3,4],[5,7]])
```

```python
# creating a dataframe using the array created
DF = pd.DataFrame(data, index=['row1','row2','row3'],columns=['col1','col2'])
# index is defining the name of the rows and names have to be stored  inside the list and same goes for columns too.
# this DataFrame is being stored in the variable DF and hence when printing then it will show the DataFrame.
```

```python
# showing the DataFrame and don't PRINT it
DF
```

# 2:- USING DICTIONARIES

```python
# using list to store data
states = ['NewYork','LosAngeles','California','LasVegas']
police_depart = ['NYPD','LAPD','CAPD','LVPD']
population = [123456,7891011,12131415,161718119]
```

```python
# using dictiionary to map each list in the form of key-pair vlaues
my_dict = {'STATES':states,'DEPARTMENT':police_depart,'POPULATION':population}
```

```python
# creating the DataFrame using the dictionary created for the lists which was mapped andn store in the variable DF
DF = pd.DataFrame(my_dict)
```

```python
#showing the Dataframe
DF
```
